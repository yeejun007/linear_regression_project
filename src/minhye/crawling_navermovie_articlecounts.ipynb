### 개봉전 1주일 기사건수 정리

```python
titles = []
before_numbers = []

for title, ds, de in zip(datas['영화명'], datas['before_1week'], datas['oneday_before ']):
    titles.append(title)   
    url = 'https://search.naver.com/search.naver?where=news&query={}&sm=tab_opt&sort=0&photo=0&field=0&reporter_article=&pd=3&ds={}&de={}'.format(title, ds, de)
    response = requests.get(url)
    dom = BeautifulSoup(response.content, 'html.parser')
    before_number = dom.select_one('#main_pack > div > div.section_head > div.title_desc.all_my > span')
    try:
        before_numbers.append(before_number.text.split("/")[1].strip())
    except:
        before_numbers.append(0)

before_df = pd.DataFrame({"영화명": titles, "before": before_numbers})
before_df.tail()
```

### 개봉후 1주일 기사건수 정리

```python
titles = []
after_numbers = []

for title, ds, de in zip(datas['영화명'], datas['release'], datas['after_1week']):
    titles.append(title)   
    url = 'https://search.naver.com/search.naver?where=news&query={}&sm=tab_opt&sort=0&photo=0&field=0&reporter_article=&pd=3&ds={}&de={}'.format(title, ds, de)
    response = requests.get(url)
    dom = BeautifulSoup(response.content, 'html.parser')
    after_number = dom.select_one('#main_pack > div > div.section_head > div.title_desc.all_my > span')
    try:
        after_numbers.append(after_number.text.split("/")[1].strip())
    except:
        after_numbers.append(0)

after_df = pd.DataFrame({"영화명": titles, "after": after_numbers})
after_df.tail()
```

```python
article_df = pd.merge(before_df, after_df)
article_df.tail()
```
