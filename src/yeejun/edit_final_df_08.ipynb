```python
date_list = []

for i, row in final_df.iterrows():
        date_list.append(row['release_date'][-4:])
```

```python
# for date in date_list:
#     if len(date) != 4:
#         print(date)
```

```python
# 감독파워 측정함수(compare_df_sep)

index = 0
director_count_list = []
movie_count_list = []

def director_func(director):
    global index
    
    total_audience = 0
    movie_num = 0
    director_list = director.split(",")
    limit_date = int(date_list[index]) - 3
    
    if len(director_list) > 1:
        director_count_list.append(len(director_list))
        for director_name in director_list:
            for i, row in compare_df_sep.iterrows():
                if director_name in row['감독'].split(","):
                    if int(date_list[index]) > int(row['개봉일'][:4]) >= limit_date:
                        total_audience += row['전국관객수']
                        movie_num += 1
    else: 
        director_count_list.append(1)
        for i, row in compare_df_sep.iterrows():
            if director_list[0] in row['감독'].split(","):
                if int(date_list[index]) > int(row['개봉일'][:4]) >= limit_date:
                    total_audience += row['전국관객수']
                    movie_num += 1
            
    index += 1
    movie_count_list.append(movie_num)
    
    return total_audience
            
        
```

```python
final_df['director_power'] = final_df['director'].apply(director_func)
```

```python
final_df['director_count'] = director_count_list
```

```python
final_df['director_movie_count'] = movie_count_list
```

```python
final_df.to_csv('final_df_10.csv')
```

```python
# 배우 성과측정 함수 (compare_df2_sep)

index = 0
actor_involved_movie_count_list = []

def actor_func(actors):
    global index
    
    total_audience = 0
    movie_count = 0
    actor_list = actors.split(",")
    limit_date = int(date_list[index]) - 3
    
    for each_actor in actor_list:
        for i, row in compare_df2_sep.iterrows():
            if each_actor in row['배우'].split(","):
                if int(date_list[index]) > int(row['개봉일'][:4]) >= limit_date:
                    total_audience += row['전국관객수']
                    movie_count += 1
        
    index += 1
    actor_involved_movie_count_list.append(movie_count)
    
    return total_audience
    
```

```python
final_df['actor_power'] = final_df['starring_actors'].apply(actor_func)
```

```python
final_df['actor_movie_count'] = actor_involved_movie_count_list
```
